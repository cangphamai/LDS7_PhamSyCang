```r
n = as.integer(readline("nhap n:"))
```

```r
A1 = matrix(runif(n*n, min=1, max=n*n), n, n)
print(A1)
```

```r
m = as.integer(readline("nhap m:"))
```

```r
A2 = matrix(runif(n*m, min=1, max=n*m), n, m)
```

```r
print(A2)
```

```r
M = solve(A1,A2)
print(M)
```

```r
print(M[1,])
```

```r
print(M[,1])
```

```r
if(n>=3 && m>=2){
    print(M[3,2])
}
```
