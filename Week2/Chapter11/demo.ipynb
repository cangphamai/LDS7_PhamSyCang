```r
x = c(46,37,35,60)
labels = c("10A1", "10A2", "10A3", "10A4")
pie(x, labels)
dev.off()
```

```r
pie(x, labels, main="class pie chart", col = rainbow(length(x)), clockwise=TRUE)
dev.off()
```

```r
piepercent = round(100*x/sum(x), 1)
pie(x, labels = paste(piepercent, "%"), main="class pie chart", col = rainbow(length(x)), clockwise=TRUE)
legend("topright", labels,cex =0.8, fill = rainbow(length(x)))
dev.off()
```

```r
H = c(5,15,20,5)
M = c("Excellent", "very good", "good", "medium")
barplot(H, names.arg = M, xlab = "Grade", ylab="Students", col = c("#9933cc"),
       main="Grade of class", border= "red")
dev.off()
```

```r
colors = c("green", "orange", "brown", "red")
classes = c("10a1", "10a2", "10a3")
grades = c("excellent", "very good", "good", "medium")
Values = matrix(c(5,6,20, 30, 40, 50, 2,9,30, 1,5,10),nrow =4, ncol=3, byrow=TRUE)
print(Values)
barplot(Values, main = "Grade of 3 class", names.arg= classes, xlab="class", ylab="student", col= colors)
legend("topleft", inset=0.1, grades, cex=0.9, fill=colors)
dev.off()
```

```r
bears = c(8,52,65,120, 156)
plot(bears, type="o", col = "red", xlab="year", ylab="numbers", main="number of bears")
dev.off()
```

```r
plot(bears, type="o", col = "red", xaxt = "n", xlab="year", ylab="numbers", main="number of bears")
axis(1, at=1:6, labels=c("2017", "2018", "2019", "2020", "2021", "2020"))
dev.off()
```

```r
dolphins =c(150, 55, 32, 11, 6,1)
whales = c(80, 54,100, 76,93, 72)
g_range = range(0, bears, dolphins, whales)
plot(bears, type="o", col = "red", xlab="year", ylab="numbers",  xaxt = "n", main="number of bears")
axis(1, at=1:5, labels=c("2017", "2018", "2019", "2020", "2021"))
lines(dolphins, type="o", col="blue")
lines(whales, type="o", col="green")
legend("topleft", c("bears", "dolphins", "whales"), cex=0.8, fill = c("red", "blue", "green"))
dev.off()
```
