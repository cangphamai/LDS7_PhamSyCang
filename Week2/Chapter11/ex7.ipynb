```r
year = c(2014,2015,2016,2017)
first_fallen= c(309,271,263,255)
one_hundred_fallen = c(309,314, 312, 304)
```

```r
plot(first_fallen, type="o", col = "red", xlab="first fallen",  xaxt = "n", main="number of fallen")
axis(1, at=1:4, labels=year)
lines(one_hundred_fallen, type="o", col="green")
legend("topleft", c("first_fallen", "one_hundred_fallen"), cex=0.8, fill = c("red", "green"))
dev.off()
```
