```r
library(plotrix)
```

```r
Tree_species = c("Beech", "Chestnut", "Hawthorn", "Red maple", "Witch Hazel", "Ye;;pw Birch")
number_of_trees = c(2,2,1,4,1,1)
piepercent = round(100*number_of_trees/sum(number_of_trees), 1)
pie3D(number_of_trees, labels = piepercent, main="class pie chart", col = rainbow(length(number_of_trees)), explode=0.1)
legend("topright", Tree_species,cex =0.8, fill = rainbow(length(number_of_trees)))
dev.off()
```

```r
barplot(number_of_trees, names.arg = Tree_species, xlab = "tree species", ylab="number", col = c("#9933cc"),
       main="number of tree", border= "red")
dev.off()
```
