```r
df = read.csv("life_batteries.txt", sep="\t", header=F)
print(df)
```

```r
data= c(df$V1, df$V2, df$V3, df$V4)
data
```

```r
par(mfrow=c(1,2))
hist(data, main = "Life batteries", xlab="Hours", xlim =c(110, max(data)+10), ylim=c(0,20), col="orange", border="blue", breaks=5)
hist(data, main = "Life batteries", xlab="Hours", xlim =c(110, max(data)+10), col="orange", border="blue", breaks=10, las=1, freq=FALSE)
lines(density(data))
```

```r
summary(data)
```

```r
var(data)
```

```r
sd(data)
```

```r
install.packages("moments")
```

```r
library("moments")
```

```r
skewness(data)
```

```r
kurtosis(data)
```
