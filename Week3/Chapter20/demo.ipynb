```r
wss = (nrow(iris) -1) * sum(apply(iris[,3:4], 2, var))
print(wss)
```

```r
for(i in 1:6) wss[i] = sum(kmeans(iris[,3:4], centers=i, nstart =20)$withinss)
```

```r
print(wss)
```

```r
plot(1:6, wss, type='b', xlab="number of clusters", ylab="wss")
```
